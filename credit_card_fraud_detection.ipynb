Under-Sampling

Build sample dataset containing similar distribution of normal transaction and fraudulent transaction

number of fraudulent transaction --> 492


```python
legit_sample = legit.sample(n=492) #provides random sample 
```

concatinating two DataFrame

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)

#axis = 0 specifies it should be added row wise
```

```python
new_dataset.head(5)
```

```python
new_dataset.tail(5)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting data into features & targets

Targets are nothing but 0 & 1

Features are this v1,v2,...,v28,amount

```python
x = new_dataset.drop(columns='Class',axis=1)  #droping column so axis is set to 1
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Now we will split this features and targets into training data and test data

For this we will use train test split which is imported from sklearn.model_selection

we need to mention four variables...

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)

#test size represents the amount of testing data we want
#so 80% of data goes to training data and 20% goes to testing data
#stratify evenly distributes y into training and testing data
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

so here we can see that in x there were 984 values and out of that 787 values gone to x_train and 197 values goes to x_test

**Model Training**

Logistic Regression

so generally we use logistic regression for binary classification problems

```python
model = LogisticRegression()
```

```python
# Training logistic regression model with training data
model.fit(x_train,y_train) 
#so as we know x_train contains all the features of training data and y_train contains corresponding labels
#labels are nothing but o & 1
```

Model Evaluation

Accuracy Score

```python
# first we are finding accuracy score on our training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on Training Data : ',training_data_accuracy)
```

```python
# we got accuracy score of 94.1% which is very good accuracy .
# so 94.1% means out of 100 predictions our model can predict almost 94 predictions correct
```

```python
# testing data is the values which our model has not seen earlier so based on that also evaluation is important
```

```python
# Accuracy on Test Data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on Testing Data : ',test_data_accuracy)
```

```python
# here we are getting almost same accuracy as of training data 
# if it would be different then our model would have been overfitted or underfitted 
# let say if accuracy of training data is 94% and accuracy of test data is 50% then it means our model has overfitted with the training data
# let say if accuracy of training data is 50% and accuracy of test data is 94% then it means model is underfit with the training data
```
